```python
x =df.iloc[0:,4:5]
x
```

```python
kmeans = KMeans(4)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
df_scaled = pd.DataFrame(df_scaled, columns=['penduduk_miskin','nama_provinsi'])
```

```python
data_with_clusters = df.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['penduduk_miskin'],data_with_clusters['nama_provinsi'],c=data_with_clusters['Clusters'],cmap='rainbow')
```

```python
wcss=[]
for i in range(1,10):
    kmeans = KMeans(i)
kmeans.fit(x)
wcss_iter = kmeans.inertia_
wcss.append(wcss_iter)

number_clusters = range(10,11)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```
